```python
import numpy as np
import pandas as pd

import random

import sys
sys.path.append("../")

from src.object import *
from src.create import *

from config.formation import *
```

```python
def apply_points(row):
    return row.win*3+row.row
```

```python
class League:
    def __init__(self, name, teams):
        self.name = name
        self.teams = teams
```

```python
df_name = pd.read_csv("../data/csv/NationalNames.csv")
df_name_list = list(df_name[df_name.Gender=='M'].Name.sample(500))
```

```python
df_place = pd.read_csv("../data/csv/results.csv")
df_place = df_place.city.unique()
```

```python
league_team = []

for i in range(20):
    min_rate = np.random.randint(70, 80)
    max_rate = min_rate + 15
    players = random_create_players(27, 3, min_rate, max_rate, df_name_list)
    formation = random_create_formation()
    A = Team(name=df_place[i], formation=formation)

    A.set_all_players(players)
    A.set_starting_players()
    A.formation.cal_team_rate()
    
    league_team.append(A)

L1 = League(name="premia", teams=league_team)
```

```python
all_team_name = [s.name for s in league_team]
output = pd.DataFrame(np.zeros((20, 5)), index=all_team_name, columns=["win", "lose", "row", "得点", "失点"], dtype=np.int8)

for i in range(20):
    for j in range(20):
        if i!= j:
            game = Game(home=league_team[i], away=league_team[j])
            game.battle()
            if game.result=="home":
                output.loc[league_team[i].name, "win"] += 1
                output.loc[league_team[j].name, "lose"] += 1
            elif game.result=="away":
                output.loc[league_team[j].name, "win"] += 1
                output.loc[league_team[i].name, "lose"] += 1
            else:
                output.loc[league_team[i].name, "row"] += 1
                output.loc[league_team[j].name, "row"] += 1
            
            output.loc[league_team[i].name, "得点"] += game.home_goal
            output.loc[league_team[i].name, "失点"] += game.away_goal
            
            output.loc[league_team[j].name, "得点"] += game.away_goal
            output.loc[league_team[j].name, "失点"] += game.home_goal
```

```python
output["得失点差"] = output["得点"]-output["失点"]
output["Points"] = output.apply(apply_points, axis=1)
output = output.sort_values("Points", ascending=False)
```
