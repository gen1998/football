```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 80)

import random

import sys
sys.path.append("../")

from src.object import *
from src.create import *

from config.formation import *
```

```python
num = 80
teams = [f"{i}" for i in range(num)]

for max_p in range(100):
    if num < pow(2, max_p):
        break

random.shuffle(teams)

result = pd.DataFrame(columns=["チームA", "チームB", "勝利", "スコア", "ラウンド"])
round_ = 1
while True:
    buff_teams = teams.copy()
    if round_==1:
        teams = teams[:(80-pow(2, max_p-1))*2]
    for i in range(0, len(teams), 2):
        game_team = teams[i:i+2]
        if len(game_team) < 2:
            continue

        if np.random.normal(0,1)>0:
            win = game_team[0]
            lose = game_team[1]
        else:
            win = game_team[1]
            lose = game_team[0]
        
        if max_p-3 >= round_:
            round_name = f"{round_}回線"
        elif max_p-2 == round_:
            round_name = f"準々決勝"
        elif max_p-1 == round_:
            round_name = f"準決勝"
        elif max_p == round_:
            round_name = f"決勝"

        output = pd.Series([game_team[0], game_team[1], win, "1-0", round_name], index=["チームA", "チームB", "勝利", "スコア", "ラウンド"])
        result.loc[f"{win}-{lose}"] = output
        buff_teams.remove(lose)
    round_ += 1
    teams = buff_teams
    if len(teams)<2:
        break
```
