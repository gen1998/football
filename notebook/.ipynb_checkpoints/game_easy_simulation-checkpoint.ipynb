```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("../data/excel/simulation.xlsx", index_col=0)
```

```python
team = list(data.index)
```

```python
mid_rate = 0.5
gk_rate = 0.2

output = pd.DataFrame([[""]*len(team)]*len(team), index=team, columns=team)
attacking = pd.DataFrame([[""]*len(team)]*len(team), index=team, columns=team)

for a_name in team:
    for b_name in team:
        if a_name == b_name:
            continue
        a_att = data.loc[a_name, "ATT"]
        a_mid = data.loc[a_name, "MID"]
        a_def = data.loc[a_name, "DEF"]
        a_gk = data.loc[a_name, "GK"]
        a_all = data.loc[a_name, "ALL"]
        
        b_att = data.loc[b_name, "ATT"]
        b_mid = data.loc[b_name, "MID"]
        b_def = data.loc[b_name, "DEF"]
        b_gk = data.loc[b_name, "GK"]
        b_all = data.loc[b_name, "ALL"]
        
        a_attacking = (a_att + a_mid*mid_rate)*a_all/(a_all+b_all) - (b_def + b_mid*mid_rate)*b_all/(a_all+b_all) - b_gk*gk_rate
        b_attacking = (b_att + b_mid*mid_rate)*b_all/(a_all+b_all) - (a_def + a_mid*mid_rate)*a_all/(a_all+b_all) - a_gk*gk_rate
        
        a_win = 0
        a_lose = 0
        row = 0
        a_all_goal = 0
        b_all_goal = 0

        for i in range(100):
            a_goal = 0
            b_goal = 0
            for j in range(9):
                a_attacking_ = a_attacking+np.random.normal(0, 17)
                b_attacking_ = b_attacking+np.random.normal(0, 17)
                
                if a_attacking_>0:
                    a_goal+=1
                if b_attacking_>0:
                    b_goal+=1
            if a_goal==b_goal:
                row+=1
            elif a_goal>b_goal:
                a_win+=1
            else:
                a_lose+=1
            a_all_goal += a_goal
            b_all_goal += b_goal

        output.loc[a_name, b_name] = f"{a_win}-{a_lose}-{row},{a_all_goal/100},{b_all_goal/100}"
        attacking.loc[a_name, b_name] = f"{a_attacking}-{b_attacking}"
```
